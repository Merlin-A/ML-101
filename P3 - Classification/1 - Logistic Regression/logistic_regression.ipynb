### Classification - To Classify between two options, Write or Wrong, One or Two, Malignant or Benign

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
dataset = pd.read_csv('./Social_Network_Ads.csv')
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values

```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state=0)
```

### Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
print(X_train, '\n\n')
print(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier.fit(X_train, y_train)
```

```python
print(classifier.predict([[28, 800000]]))
```

```python
model = classifier.predict(X_train)
```

```python
print(model)
```

```python
sum = 0
for i in range (len(model)):
    print(y_train[i], ' -- ', model[i])
    if (y_train[i] == model[i]):
        sum = sum + 1
    
print("Accuracy -" ,sum/len(model)*100)
```

### Prediciting Test set Results


```python
y_model = classifier.predict(X_test)
np.set_printoptions(precision=2)
print(np.concatenate((y_model.reshape(len(y_model), 1), y_test.reshape(len(y_test), 1)), 1))
```

### Confusion Matrix 

#### Train Plot

```python
# Confusion Matrix is used to calculate and classify how accurate the model was in predicting the results
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, accuracy_score
cMatrix = confusion_matrix(y_train, model)
display = ConfusionMatrixDisplay(cMatrix)
display.plot()
accuracy_score(y_train, model)
```

#### Test Plot

```python
print(accuracy_score(y_test, y_model))
disp = ConfusionMatrixDisplay(confusion_matrix(y_test, y_model))
disp.plot()
```

### Visualizing Train Results

```python
from matplotlib.colors import ListedColormap
X_set, y_set = sc.inverse_transform(X_train), y_train
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 10, stop = X_set[:, 0].max() + 10, step = 0.25),
                     np.arange(start = X_set[:, 1].min() - 1000, stop = X_set[:, 1].max() + 1000, step = 0.25))
plt.contourf(X1, X2, classifier.predict(sc.transform(np.array([X1.ravel(), X2.ravel()]).T)).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1], c = ListedColormap(('red', 'green'))(i), label = j)
plt.title('Logistic Regression (Test set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```
