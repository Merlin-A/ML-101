```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('./Social_Network_Ads.csv')
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0, test_size=0.25)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier(criterion='entropy', random_state=0)
classifier.fit(X_train, y_train)
```

### Test Visualizing Data

```python
model = classifier.predict(X_test)
print(np.concatenate((model.reshape(len(model),1), y_test.reshape(len(y_test), 1)),1))

from sklearn.metrics import ConfusionMatrixDisplay, confusion_matrix, accuracy_score
disp = ConfusionMatrixDisplay(confusion_matrix(y_test, model))
disp.plot()
print("Accuracy - ", accuracy_score(y_test, model))
```

```python
## Overfitting
model = classifier.predict(X_train)
print(np.concatenate((model.reshape(len(model),1), y_train.reshape(len(y_train), 1)),1))

from sklearn.metrics import ConfusionMatrixDisplay, confusion_matrix, accuracy_score
disp = ConfusionMatrixDisplay(confusion_matrix(y_train, model))
disp.plot()
print("Accuracy - ", accuracy_score(y_train, model))
```
