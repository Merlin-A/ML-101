## Clustering 
#### Clustering is a unsupervised learning algorithm
#### It is used to analyze random data and then form classes/clusters from them 
#### It can then be used to determine the class/cluster of some new data entry


```python
import pandas as pd  
import numpy as np
import matplotlib.pyplot as plt
```

##### There is no dependent variable in a supervised learning algorithm that can be directly used to predict the classes

```python
dataset = pd.read_csv('./Mall_Customers.csv')
X = dataset.iloc[:, 3:].values # Just use the two major independent variables to achieve the best result 

print(X)
```

## Use Elbow method to find the optimal number of clusters
### WCSS is the sum of squared distances of each data point in all clusters to their respective centroids.|
###### As we need to determine the optimal number of clusters we use the elbow method

```python
# HERE WE TRAIN THE MODEL USING KMEANS TO FIND THE OPTIMAL NUMBER OF CLUSTERS, WE KEEP THE CLUSTER RANGE BETN 1 - 10
# AND ONCE THE WCSS CHART STARTS GOING SOUTH WE TAKE THAT POINT AS THE OPTIMAL VALUE
from sklearn.cluster import KMeans
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init='k-means++', random_state=0) # KMEANS++ IS USED TO AVOID THE RANDOM INITIALIZATION TRAP 
    kmeans.fit(X)
    wcss.append(kmeans.inertia_) # HERE WE ADD THE DATA WE GET THROUGH THE ALGO AND ADD IT TO WCSS ARRAY
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('No. of Clusters')
plt.ylabel('WCSS')
plt.show()
    
```

## Training Model
##### The independent variable x is used to determine clusters of new data entries to the model

```python
kmeans = KMeans(n_clusters = 5, init='k-means++', random_state=0)
y_kmeans = kmeans.fit_predict(X)
print(y_kmeans.reshape(len(y_kmeans), 1))
```

## Visualize Clusters

#### Thus in Clustering, we find some random data and try to find pattern in them and categorizing them into certain groups, so as to normalize the data

```python
plt.scatter(X[y_kmeans == 0,0], X[y_kmeans == 0, 1], s = 100, c = 'red', label = 'Cluster 1')
plt.scatter(X[y_kmeans == 1,0], X[y_kmeans == 1, 1], s = 100, c = 'blue', label = 'Cluster 2')
plt.scatter(X[y_kmeans == 2,0], X[y_kmeans == 2, 1], s = 100, c = 'green', label = 'Cluster 3')
plt.scatter(X[y_kmeans == 3,0], X[y_kmeans == 3, 1], s = 100, c = 'cyan', label = 'Cluster 4')
plt.scatter(X[y_kmeans == 4,0], X[y_kmeans == 4, 1], s = 100, c = 'black', label = 'Cluster 5')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 190, c = 'orange', label = 'Centroids') # TO GET CENTROID OF THE CLUS
plt.title('Clustering Graph')
plt.xlabel('Annual Income')
plt.ylabel('Score')
plt.legend()
plt.show()
```
